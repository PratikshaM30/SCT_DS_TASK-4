```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

file_path = "/content/Traffic_Accident_Analysis_Clean_Data.xlsx"
df = pd.read_excel(file_path)

print("First 5 rows of dataset:")
display(df.head())

print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
display(df.describe())

print("\nMissing Values:")
print(df.isnull().sum())

plt.figure(figsize=(6,4))
df['Severity'].value_counts().plot(kind='bar')
plt.xlabel("Severity Level")
plt.ylabel("Number of Accidents")
plt.title("Accident Severity Distribution")
plt.show()

plt.figure(figsize=(7,4))
df['Vehicle_Type'].value_counts().plot(kind='bar')
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Vehicle Type")
plt.show()

plt.figure(figsize=(7,4))
df['Weather_Condition'].value_counts().plot(kind='bar')
plt.xlabel("Weather Condition")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Weather Condition")
plt.show()

plt.figure(figsize=(7,4))
df['Road_Condition'].value_counts().plot(kind='bar')
plt.xlabel("Road Condition")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Road Condition")
plt.show()

total_injuries = df['Injuries'].sum()
total_fatalities = df['Fatalities'].sum()

print("Total Injuries:", total_injuries)
print("Total Fatalities:", total_fatalities)

plt.figure(figsize=(6,4))
sns.boxplot(x='Severity', y='Injuries', data=df)
plt.title("Severity vs Injuries")
plt.show()

plt.figure(figsize=(6,4))
df['Number_of_Vehicles'].value_counts().sort_index().plot(kind='bar')
plt.xlabel("Number of Vehicles")
plt.ylabel("Accident Count")
plt.title("Accidents by Number of Vehicles Involved")
plt.show()

df.to_excel("Traffic_Accident_Analysis_Final.xlsx", index=False)
print("Final cleaned file saved successfully!")
```
